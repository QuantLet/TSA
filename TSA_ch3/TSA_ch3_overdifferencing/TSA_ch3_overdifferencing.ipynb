```python
import osimport numpy as npimport pandas as pdimport matplotlib.pyplot as pltfrom statsmodels.tsa.stattools import acf, pacf, adfuller, kpssimport warningswarnings.filterwarnings('ignore')COLORS = {'blue': '#1A3A6E', 'red': '#DC3545', 'green': '#2E7D32', 'orange': '#E67E22', 'gray': '#666666', 'purple': '#8E44AD'}BLUE, RED, GREEN, ORANGE, GRAY, PURPLE = COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['orange'], COLORS['gray'], COLORS['purple']plt.rcParams.update({    'figure.facecolor': 'none', 'axes.facecolor': 'none', 'savefig.facecolor': 'none',    'savefig.transparent': True, 'axes.spines.top': False, 'axes.spines.right': False,    'axes.grid': False, 'font.size': 10, 'axes.titlesize': 12, 'axes.labelsize': 10,    'xtick.labelsize': 9, 'ytick.labelsize': 9, 'legend.fontsize': 9, 'figure.dpi': 150,    'lines.linewidth': 1.2, 'axes.linewidth': 0.6, 'legend.facecolor': 'none',    'legend.framealpha': 0, 'legend.edgecolor': 'none',})np.random.seed(42)def save_chart(fig, name):    fig.savefig(f'{name}.pdf', bbox_inches='tight', transparent=True, dpi=150)    fig.savefig(f'{name}.png', bbox_inches='tight', transparent=True, dpi=150)    print(f'Saved: {name}')
```

```python
# Overdifferencing: what happens when we difference too muchn = 500np.random.seed(42)# Generate I(1) process (random walk)eps = np.random.normal(0, 1, n)rw = np.cumsum(eps)# First difference (correct for I(1))diff1 = np.diff(rw)# Second difference (overdifferenced!)diff2 = np.diff(diff1)fig, axes = plt.subplots(2, 3, figsize=(12, 6))# Top row: the three seriesaxes[0, 0].plot(rw, color=BLUE, linewidth=0.8)axes[0, 0].set_title('Original $Y_t$ (Random Walk)', fontweight='bold')axes[0, 0].set_ylabel(r'$Y_t$')axes[0, 1].plot(diff1, color=GREEN, linewidth=0.8)axes[0, 1].set_title(r'$\Delta Y_t$ (Correct)', fontweight='bold')axes[0, 1].set_ylabel(r'$\Delta Y_t$')axes[0, 2].plot(diff2, color=RED, linewidth=0.8)axes[0, 2].set_title(r'$\Delta^2 Y_t$ (Overdifferenced!)', fontweight='bold')axes[0, 2].set_ylabel(r'$\Delta^2 Y_t$')# Bottom row: ACF of eachfor i, (series, color, title) in enumerate([    (rw, BLUE, 'ACF: Original'),    (diff1, GREEN, 'ACF: $\\Delta Y_t$'),    (diff2, RED, 'ACF: $\\Delta^2 Y_t$')]):    acf_vals = acf(series, nlags=20)    axes[1, i].bar(range(len(acf_vals)), acf_vals, color=color, alpha=0.7, width=0.6)    ci = 1.96 / np.sqrt(len(series))    axes[1, i].axhline(ci, color=GRAY, linestyle='--', linewidth=0.8)    axes[1, i].axhline(-ci, color=GRAY, linestyle='--', linewidth=0.8)    axes[1, i].axhline(0, color='black', linewidth=0.3)    axes[1, i].set_title(title, fontweight='bold')    axes[1, i].set_xlabel('Lag')    if i == 0:        axes[1, i].set_ylabel('ACF')# Annotate the overdifferencing signatureacf_od = acf(diff2, nlags=20)axes[1, 2].annotate(f'Lag 1 = {acf_od[1]:.2f}\n(overdiff. signal)',                     xy=(1, acf_od[1]), xytext=(6, acf_od[1] - 0.15),                     arrowprops=dict(arrowstyle='->', color=RED),                     fontsize=9, color=RED, fontweight='bold')plt.tight_layout()save_chart(fig, 'sem3_overdifferencing')plt.show()
```
