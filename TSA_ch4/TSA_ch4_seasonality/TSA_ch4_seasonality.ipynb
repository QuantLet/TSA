```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import acf, pacf
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
COLORS = {'blue': '#1A3A6E', 'red': '#DC3545', 'green': '#2E7D32', 'orange': '#E67E22', 'gray': '#666666', 'purple': '#8E44AD'}
BLUE, RED, GREEN, ORANGE, GRAY, PURPLE = COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['orange'], COLORS['gray'], COLORS['purple']

plt.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none', 'savefig.facecolor': 'none',
    'savefig.transparent': True, 'axes.spines.top': False, 'axes.spines.right': False,
    'axes.grid': False, 'font.size': 10, 'axes.titlesize': 12, 'axes.labelsize': 10,
    'xtick.labelsize': 9, 'ytick.labelsize': 9, 'legend.fontsize': 9, 'figure.dpi': 150,
    'lines.linewidth': 1.2, 'axes.linewidth': 0.6, 'legend.facecolor': 'none',
    'legend.framealpha': 0, 'legend.edgecolor': 'none',
})

def save_chart(fig, name):
    fig.savefig(f'{name}.pdf', bbox_inches='tight', transparent=True, dpi=150)
    fig.savefig(f'{name}.png', bbox_inches='tight', transparent=True, dpi=150)
    print(f'Saved: {name}')
```

```python
np.random.seed(7)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Monthly data with s=12
t1 = np.arange(60)
d1 = pd.date_range('2018-01-01', periods=60, freq='MS')
axes[0].plot(d1, 50 + 0.2*t1 + 10*np.sin(2*np.pi*t1/12) + np.random.normal(0, 2, 60),
             color=BLUE, lw=1.2)
axes[0].axvspan(d1[12], d1[23], color=ORANGE, alpha=0.15, label='One cycle (s=12)')
axes[0].set_title('Monthly: s = 12', fontweight='bold')
axes[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), frameon=False)

# Quarterly data with s=4
t2 = np.arange(40)
d2 = pd.date_range('2014-01-01', periods=40, freq='QS')
axes[1].plot(d2, 30 + 0.3*t2 + 8*np.sin(2*np.pi*t2/4) + np.random.normal(0, 1.5, 40),
             color=GREEN, lw=1.2)
axes[1].axvspan(d2[4], d2[7], color=PURPLE, alpha=0.15, label='One cycle (s=4)')
axes[1].set_title('Quarterly: s = 4', fontweight='bold')
axes[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), frameon=False)

plt.tight_layout(); save_chart(fig, 'ch4_seasonality'); plt.show()
```
