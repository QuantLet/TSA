```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import acf, pacf
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
COLORS = {'blue': '#1A3A6E', 'red': '#DC3545', 'green': '#2E7D32', 'orange': '#E67E22', 'gray': '#666666', 'purple': '#8E44AD'}
BLUE, RED, GREEN, ORANGE, GRAY, PURPLE = COLORS['blue'], COLORS['red'], COLORS['green'], COLORS['orange'], COLORS['gray'], COLORS['purple']

plt.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none', 'savefig.facecolor': 'none',
    'savefig.transparent': True, 'axes.spines.top': False, 'axes.spines.right': False,
    'axes.grid': False, 'font.size': 10, 'axes.titlesize': 12, 'axes.labelsize': 10,
    'xtick.labelsize': 9, 'ytick.labelsize': 9, 'legend.fontsize': 9, 'figure.dpi': 150,
    'lines.linewidth': 1.2, 'axes.linewidth': 0.6, 'legend.facecolor': 'none',
    'legend.framealpha': 0, 'legend.edgecolor': 'none',
})

def save_chart(fig, name):
    fig.savefig(f'{name}.pdf', bbox_inches='tight', transparent=True, dpi=150)
    fig.savefig(f'{name}.png', bbox_inches='tight', transparent=True, dpi=150)
    print(f'Saved: {name}')
```

```python
# Load airline passengers
data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv',
                    parse_dates=['Month'], index_col='Month')
passengers = data['Passengers']
passengers.index.freq = 'MS'
log_passengers = np.log(passengers)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 7))

# Original log-transformed
axes[0,0].plot(log_passengers.index, log_passengers.values, color=BLUE, lw=1.2)
axes[0,0].set_title('$\\log(Y_t)$: Original (log-transformed)', fontweight='bold')
axes[0,0].set_ylabel('$\\log$ Passengers')

# Regular difference
d1 = log_passengers.diff().dropna()
axes[0,1].plot(d1.index, d1.values, color=ORANGE, lw=1)
axes[0,1].axhline(0, color=GRAY, ls='--')
axes[0,1].set_title('$\\Delta \\log(Y_t)$: Trend Removed', fontweight='bold')
axes[0,1].set_ylabel('Regular Diff')

# Seasonal difference
d12 = log_passengers.diff(12).dropna()
axes[1,0].plot(d12.index, d12.values, color=GREEN, lw=1)
axes[1,0].axhline(0, color=GRAY, ls='--')
axes[1,0].set_title('$\\Delta_{12} \\log(Y_t)$: Seasonality Removed', fontweight='bold')
axes[1,0].set_ylabel('Seasonal Diff')

# Both differences
db = log_passengers.diff().diff(12).dropna()
axes[1,1].plot(db.index, db.values, color=RED, lw=1)
axes[1,1].axhline(0, color=GRAY, ls='--')
axes[1,1].set_title('$\\Delta\\Delta_{12} \\log(Y_t)$: Stationary', fontweight='bold')
axes[1,1].set_ylabel('Both Diffs')

plt.tight_layout(); save_chart(fig, 'ch4_sarima_model'); plt.show()
```
